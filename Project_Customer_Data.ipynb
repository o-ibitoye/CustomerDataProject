### Question 6

#### The client states that there may be some issues with credit card entries. Any customers that have more than 10 years between their start and end date need writing to a separate file, called remove_ccard.json, in the JSON data format. The client will manually deal with these later based on your output. They request that you write a function to help perform this, which accepts a single row from the CSV data, and outputs whether the row should be flagged. This can then be used when determining whether to write the current person to the remove_ccard file

```python
def credit_card(user_details):
    
    """
    A function that checks the start and expiry date 
    on credit card and determine if it's more than 10 Years
    """
    
    for i in range(len(user_details)):
        breakdown_start = user_details['Credit Card']['Credit Card Start Date']
        breakdown_expiry = user_details['Credit Card'] ['Credit Card Expiry Date']
   
    
    #calculating the start month and year
        
        start_month = int(breakdown_start.split('/')[0])
        start_year = int(breakdown_start.split('/')[1])
        
    #calculating the expiry  month and year
        
        expiry_month = int(breakdown_expiry.split('/')[0])
        expiry_year = int(breakdown_expiry.split('/')[1])
        
    #calculating the difference in month and year
        month_difference = expiry_month - start_month
        year_difference = expiry_year - start_year
        
        if month_difference < 0:
            year_difference = year_difference - 1
            month_difference = month_difference + 12 
        
        return(year_difference > 10)

credit_card
```

```python
#Validate that the function works as expected, should return True if Year is > 10 or False if Year < 10

for i, row in enumerate(user_data):
    if i == 5:
        break
    else:
        print(row['Credit Card']['Credit Card Start Date'],row['Credit Card']['Credit Card Expiry Date'], credit_card(row))
```

```python
#creating remove_ccard.json file

remove_ccard = []

with open('remove_ccard.json', mode = 'w') as file:
    for user_details in user_data:
        if credit_card(user_details):
            remove_ccard.append(user_details)
    json.dump(remove_ccard, file)
```

### Question 7

#### You have been tasked with calculating some additional metrics which will be used for ranking customers. You should create a new data attribute for our customers called “Salary-Commute”. Reading in from processed.json

### Question 7a

#### Add, and calculate appropriately, this new attribute. It should represent the Salary that a customer earns, per mile of their commute.
#### i. Note: If a person travels 1 or fewer commute miles, then their salary- commute would be just their salary

```python
def salary_pm(user_details):
    
    """
    A function to calculate the Salary a customer
    earns per mile of their commute
    """
        
    if user_details['Distance Commuted to Work(miles)'] <= 1:
        salary_commute = user_details['Yearly Salary (GBP)']
    else:
        salary_commute = user_details['Yearly Salary (GBP)'] / user_details['Distance Commuted to Work(miles)']
    #converting salary_commute to a float data type
    salary_commute = float(salary_commute)
    #converting salary_commute into two decimal places
    salary_commute = f'{salary_commute:.2f}' 

    return salary_commute
```

```python
#reading in and updated processed.json file
with open('processed.json', mode='r') as file:
    new_processed_data = json.load(file)
    for user_details in new_processed_data:
        user_details['Salary Per Mile'] = salary_pm(user_details)
```

### Question 7b

#### Sort these records by that new metric, in ascending order.

```python
#sorting new processed data by Salary Per Mile
sorted_data = sorted(new_processed_data, key=lambda d: d['Salary Per Mile'])
```
